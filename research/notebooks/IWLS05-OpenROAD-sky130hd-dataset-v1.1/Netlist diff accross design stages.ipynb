# Netlist diff accross design stages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

from eda_schema.entity import PHASES
from eda_schema.dataset import Dataset
from eda_schema.db import SQLitePickleDB
from eda_schema.utils.netlist_diff import compare_netlists_by_cells, compare_netlists_by_nets

DATASET_DIR = "../../dataset/dataset_openroad_sky130hd_iwls05_v1.1"
dataset = Dataset(SQLitePickleDB(DATASET_DIR))
dataset.load_dataset(netlist_id='id-000001', load_timing_paths=False)

CIRCUITS = sorted(list(set([circuit for circuit, _, _ in dataset])))
```

## Cell change comparisons between adjacent design stages

```python
data = []
for circuit in CIRCUITS:
    row = {"circuit": circuit}
    for i in range(len(PHASES) - 1):
        phase1, phase2 = PHASES[i], PHASES[i + 1]
        netlist1 = dataset[(circuit, 'id-000001', phase1)]
        netlist2 = dataset[(circuit, 'id-000001', phase2)]
        _, _, comparison_summary = compare_netlists_by_cells(netlist1, netlist2)
        row.update({f"{phase2}_{k}": v for k, v in comparison_summary.items()})
    data.append(row)

cell_change_df = pd.DataFrame(data)

for i in range(len(PHASES) - 1):
    phase1, phase2 = PHASES[i], PHASES[i + 1]
    print(f"initial stage: {phase1}\nfinal stage: {phase2}")

    phase_df = cell_change_df[["circuit"] + [c for c in cell_change_df.columns if phase2 in c]]
    phase_df.columns = [c.replace(phase2+"_", "") for c in phase_df.columns]
    display(phase_df)
```

## Cell change comparisons of post-floorplan designs with downstream design stages

```python
data = []
for circuit in CIRCUITS:
    row = {"circuit": circuit}
    for i in range(len(PHASES) - 1):
        phase1, phase2 = "floorplan", PHASES[i + 1]
        netlist1 = dataset[(circuit, 'id-000001', phase1)]
        netlist2 = dataset[(circuit, 'id-000001', phase2)]
        _, _, comparison_summary = compare_netlists_by_cells(netlist1, netlist2)
        row.update({f"{phase2}_{k}": v for k, v in comparison_summary.items()})
    data.append(row)

cell_change_df = pd.DataFrame(data)

for i in range(len(PHASES) - 1):
    phase1, phase2 = "floorplan", PHASES[i + 1]
    print(f"initial stage: {phase1}\nfinal stage: {phase2}")

    phase_df = cell_change_df[["circuit"] + [c for c in cell_change_df.columns if phase2 in c]]
    phase_df.columns = [c.replace(phase2+"_", "") for c in phase_df.columns]
    display(phase_df)
```
